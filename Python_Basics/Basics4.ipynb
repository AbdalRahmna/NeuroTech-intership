```python
class Car:
    def __init__(self, speed, color, dist=0):
        self.speed = speed
        self.color = color
        self.dist = dist


    def move(self):
        self.dist += self.speed

    def move_with_time(self,time):
        self.dist += self.speed * time
```

```python

my_car = Car(speed=10, color='red')


print(f"Initial distance: {my_car.dist}")

# Move the car
my_car.move()

# Print distance after moving
print(f"Distance after moving: {my_car.dist}") 

# Move the car with time
my_car.move_with_time(time=2)

# Print distance after moving with time
print(f"Distance after moving with time: {my_car.dist}")  
```

## inhertitance Classes : 

```python
class person:

    def __init__(self,first,second,age):

        self.first=first
        self.second=second
        self.age=age
    def __str__(self):
        return(f"my first_name is {self.first} and second name is {self.second} & age ={self.age}")

class student(person):

    def __init__(self,class_level,first,second,age):

        person.__init__(self,first,second,age)
        self.class_level=class_level

    def __str__(self):
        return(f"class_level={self.class_level},my first_name is {self.first} and second name is {self.second} & age ={self.age}")

class teacher(person):

    def __init__(self,address,first,second,age):

        person.__init__(self,first,second,age)
        self.address=address

    def __str__(self):
       return(f"address={self.address},my first_name is {self.first} and second name is {self.second} & age={self.age}")


p=person("Ayman","Mahmoud",25)
print(p)

s=student(5,"Ahmed","Ali",21)
print(s)

t=teacher("Cairo","Ayman","Mahmoud",25)
print(t)
    
```

```python
class person:
    def __init__(self, first_name, last_name, age):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        
    
    def __str__(self):
        return f"Hi! My name is {self.first_name} {self.last_name} and my age is {self.age} "

class programmer(person):
    def __init__(self,first,second,age,programming,time):

        person.__init__(self,first,second,age)
        self.programming=programming
        self.time=time

    def __str__(self):
        return f"Hi! I am a programmer, programming in {self.programming}, and I have programmed for {self.time} hours   "

p=programmer("Abdo","Ahmed",23,"ML",10)
print(p)
```

```python
class vechial:
    def __init__(self, engine, color, type_vechial):
        self.engine = engine
        self.color = color
        self.type_vechial = type_vechial

    def display(self):
        return f"This {self.type_vechial} with {self.color} color is fantastic with its {self.engine} engine"

class Bus(vechial):
    def __init__(self, engine, color, type_vechial, speed, time):
        super().__init__(engine, color, type_vechial)
        self.speed = speed
        self.dist = self.speed * time

    def display(self):
        speed_description = "very fast" if self.speed > 120 else "slow"
        return f"This {self.type_vechial} with {self.color} color is fantastic with its {self.engine} engine and walks {self.dist}. The bus is {speed_description}."

# Instantiate the vechial class
v = vechial("G1", "Pink", "Taxi")
print(v.display())

# Instantiate the Bus class
b1 = Bus("bus1", "Yellow", "M2", 120, 10)
print(b1.display())


```

```python
# Text inside the dictonary :

class Text:
    def __init__(self, text):
        self.text = text

    def split_text(self):
        result = {}
        word = ""
        position = 1  
        for c in self.text:
            if c != ' ':
                word += c
            else:
                if word:  
                    result[position] = word
                    position += 1
                    word = ""
        if word:  
            result[position] = word
        return result

t = Text("Hello how are you")
print("Final Result :",t.split_text())
```

## operations on the list (operatoprs):

```python
def sub_list(l1,l2):
    result=[]
    for i,j in zip(l1,l2):
        result.append(i-j)
    return result   

print("Add of the 2 list --->>",sub_list([5,10,15],[100,12,13]))
```

```python
def Add_list(l1,l2):
    result=[]
    for i,j in zip(l1,l2):
        result.append(i+j)
    return result   

print(Add_list([5,10,15],[100,12,13]))
```

```python
class lst:
    def __init__(self,i):
        self.p=i

    def __add__(self,other):
        result=[]
        for i in range(len(self.p)) :
            result .append(self.p[i]+other.p[i])
        return result    

l1=lst([1,2,3,4])
l2=lst([1,2,3,4])
print("Sum of the 2 list -->>",l1+l2)
```

```python
import re

def clean_text(file_path):
    
    with open(file_path, "r+") as t:
   
      file = t.read()
    
      cleaned_text = re.sub(r'[^a-zA-Z\s]', '', file).lower()
      cleaned_text = re.sub(r"[0-9]+", " ", cleaned_text) 

      cleaned_text = re.sub(r"\w*\d\w*", " ", cleaned_text) 
      cleaned_text = re.sub(r"<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});"," ",cleaned_text)  

      cleaned_text = re.sub(r"https?://\S+|www\.\S+"," ",cleaned_text) 
    
      t.write(cleaned_text)

    return cleaned_text

path="J. K. Rowling - Harry Potter 1 - Sorcerer's Stone.txt"
print(clean_text(path))
```
