```python
data=pd.read_csv("titanic.csv")
data.head(25)
```

```python
u=data["Cabin"].unique()
u
```

```python
replacement_dict = {
                    'C85': 'C',
                    'C123':'C',
                     'C103':'C',
                     'D56':'D',
                     'A6':'A',
                     'C23':'C',
                     'C25':'C',
                     'C27':'C',
                     'B78':'B',
                     'B42':'B',
                     'C148':'C'}
data['Cabin'] = data['Cabin'].replace(replacement_dict)
data['Cabin']
```

```python
data['Cabin'] = data['Cabin'].replace(to_replace=r'\d+', value='', regex=True)
data['Cabin'] = data['Cabin'].fillna('N')
data
```

```python
def rc(value):
    if pd.isna(value['Cabin']):
        if pd.isna(value['Embarked']):
            return 'NN'
        else:
            return 'N' + value['Embarked']
    else:
        if pd.isna(value['Embarked']):
            return value['Cabin'][0] + 'N'
        else:
            return value['Cabin'][0] + value['Embarked']
```

```python
data['CE']=data[['Cabin','Embarked']].apply(rc,axis=1)
data
```

```python
g=data[(data['Age'] < 30) & (data['CE'] == 'NS')]
len(g)
```

```python
def rc_simplified(row):
    cabin = 'N' if pd.isna(row['Cabin']) else row['Cabin'][0]
    embarked = 'N' if pd.isna(row['Embarked']) else row['Embarked']
    return cabin + embarked
```
